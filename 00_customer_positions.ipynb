```python
import pandas as pd
import folium
```

```python
# Charger le fichier CSV
df = pd.read_csv('customer_details_v1.csv')
# Supprimer les lignes avec des valeurs NaN
df = df.dropna()
# Arrondir la colonne avg_freight_value à deux décimales
df['avg_freight_value'] = df['avg_freight_value'].round(2)
# Arrondir la colonne avg_price à deux décimales
df['avg_price'] = df['avg_price'].round(2)
# Arrondir la colonne avg_payment à deux décimales
df['avg_payment'] = df['avg_payment'].round(2)
# Convertir les colonnes 'first_order' et 'last_order' en format de date
df['first_order'] = pd.to_datetime(df['first_order'])
df['last_order'] = pd.to_datetime(df['last_order'])
```

```python
# Créer un nouveau DataFrame avec des positions uniques
unique_positions_df = df.drop_duplicates(subset=['geolocation_lat', 'geolocation_lng'])

# Enregistrer le nouveau DataFrame dans un fichier CSV
unique_positions_df.to_csv('unique_customer_positions.csv', index=False)
```

```python
# Calculer la moyenne des valeurs de geolocation_lat et geolocation_lng
mean_lat = unique_positions_df['geolocation_lat'].mean()
mean_lng = unique_positions_df['geolocation_lng'].mean()

# Créer une carte centrée sur la moyenne des positions
map = folium.Map(location=[mean_lat, mean_lng], zoom_start=1)

# Ajouter des marqueurs pour chaque position de client
for index, row in unique_positions_df.iterrows():
    folium.Marker(location=[row['geolocation_lat'], row['geolocation_lng']],
                  # popup=row['avg_payment']
                  ).add_to(map)

# Sauvegarder la carte dans un fichier HTML
map.save('customer_map.html')
```
